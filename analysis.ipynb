# Análisis de Herramientas de IA en Educación

Este cuaderno realiza la ingesta de los datos crudos desde Kaggle mediante la librería `kagglehub` y realiza un perfilamiento básico de la información.

```python
import kagglehub
import pandas as pd
import os

# Identificador del dataset
dataset_id = "visionlangai/ai-tools-usage-analysis-in-education"

# Descargar última versión desde Kaggle
path = kagglehub.dataset_download(dataset_id)
print("Directorio de descarga:", path)

# Suponiendo que contiene un archivo CSV, vamos a listar el contenido
archivos = os.listdir(path)
print("Archivos encontrados:", archivos)

# Cargamos el primer CSV detectado (usualmente solo hay uno)
csv_file = [f for f in archivos if f.endswith('.csv')][0]
csv_path = os.path.join(path, csv_file)

df = pd.read_csv(csv_path)
df.head()
```
